```python
import nibabel as nib
from pathlib import Path

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
subject = 'sub-OAS30001'
session = 'ses-d0129'

sub_ses_path = Path.cwd().parent / 'data' / 'data_raw' / subject / session
print(sub_ses_path)
```

```python
anat2_path = sub_ses_path / 'anat' / 'sub-OAS30001_ses-d0129_run-02_T1w.nii.gz'
t1_img = nib.load(anat2_path)
t1_metadata = t1_img.header
print(t1_metadata)
```

```python
t1_data = t1_img.get_fdata()
t1_data
```

```python
t1_data.shape
```

The 3 numbers given here represent the number of values along a respective dimension (x,y,z). This brain was scanned in 176 slices with a resolution of 256 x 256 voxels per slice. That means there are:

176 * 256 * 256 = 11.534.336 voxels in total.

## Visualizing the data

```python
slices = [t1_data[60, :, :], t1_data[:, 155, :], t1_data[:, :, 118]]

fig, axes = plt.subplots(1, len(slices))
for i, s in enumerate(slices):
    axes[i].imshow(s.T, cmap="gray", origin="lower")
```
